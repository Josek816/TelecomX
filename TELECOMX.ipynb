## **Manejo de inconsistencias**

```python
df = df.drop_duplicates()
```

```python
df['Churn'] = df['Churn'].replace('', pd.NA)
```

```python
df = df.dropna(subset=['Churn'])
```

```python
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
```

```python
df = df.dropna(subset=['account.Charges.Total'])
```

```python
cols_obj = df.select_dtypes(include='object').columns
```

```python
for col in cols_obj:
    df[col] = df[col].str.strip().str.lower().replace('no internet service', 'no').replace('no phone service', 'no')
```

```python
print("Nulos después de limpieza:\n", df.isnull().sum().sort_values(ascending=False).head())
```

## **Columna de cuentas diarias**

```python
df['Cuentas_Diarias'] = (df['account.Charges.Monthly'] / 30).round(2)
```

```python
df[['account.Charges.Monthly', 'Cuentas_Diarias']].head()
```

## **Estandarización y transformación de datos (opcional)**

```python
binarias = [
    'customer.Partner', 'customer.Dependents',
    'phone.PhoneService', 'phone.MultipleLines',
    'internet.OnlineSecurity', 'internet.OnlineBackup', 'internet.DeviceProtection',
    'internet.TechSupport', 'internet.StreamingTV', 'internet.StreamingMovies',
    'account.PaperlessBilling'
]
```

```python
for col in binarias:
    df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
df['Churn'] = df['Churn'].map({'yes': 1, 'no': 0})
```

```python
df.rename(columns={
    'customer.gender': 'Genero',
    'customer.SeniorCitizen': 'AdultoMayor',
    'customer.Partner': 'TienePareja',
    'customer.Dependents': 'TieneDependientes',
    'customer.tenure': 'MesesCliente',
    'account.Charges.Monthly': 'FacturaMensual',
    'account.Charges.Total': 'FacturaTotal',
    'Cuentas_Diarias': 'FacturaDiaria'
}, inplace=True)
```

## **Distribución de evasión**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Distribución de Churn (Evasión de clientes)')
plt.xticks([0, 1], ['No se dio de baja', 'Se dio de baja'])
plt.xlabel('Churn')
plt.ylabel('Cantidad de clientes')
plt.tight_layout()
plt.show()
```

## **Recuento de evasión por variables categóricas**

```python
def churn_por_categoria(columna):
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=columna, hue='Churn', palette='Set2')
    plt.title(f'Evasión según {columna}')
    plt.ylabel('Cantidad')
    plt.xlabel(columna)
    plt.legend(['No se dio de baja', 'Se dio de baja'])
    plt.tight_layout()
    plt.show()
```

```python
categoricas = ['Genero', 'account.Contract', 'account.PaymentMethod', 'AdultoMayor', 'TienePareja', 'TieneDependientes']
```

```python
for col in categoricas:
    churn_por_categoria(col)
```

## **Conteo de evasión por variables numéricas**

```python
def grafico_distribucion_numerica(columna):
    plt.figure(figsize=(6,4))
    sns.kdeplot(data=df, x=columna, hue='Churn', fill=True, palette='Set2', common_norm=False, alpha=0.5)
    plt.title(f'Distribución de {columna} según evasión')
    plt.xlabel(columna)
    plt.ylabel('Densidad')
    plt.tight_layout()
    plt.show()
```

```python
numericas = ['MesesCliente', 'FacturaTotal', 'FacturaMensual', 'FacturaDiaria']

for col in numericas:
    grafico_distribucion_numerica(col)
```

---

## 🔹 Introducción

En este proyecto se abordó el problema de **evasión de clientes (Churn)** para la empresa **TelecomX**, dedicada a ofrecer servicios de telefonía, internet y entretenimiento.  
El objetivo principal fue **identificar patrones y variables asociadas a la cancelación del servicio**, para así generar insights que permitan **reducir la pérdida de clientes y mejorar la retención**.

---

## 🔹 Limpieza y Tratamiento de Datos

Para asegurar un análisis confiable, se aplicaron los siguientes pasos:

- ✅ **Carga de datos** desde archivo JSON y conversión a DataFrame.
- ✅ **Eliminación de duplicados y valores faltantes** (especialmente en `Churn` y `FacturaTotal`).
- ✅ **Conversión de columnas a tipos apropiados** (por ejemplo, cargos de texto a numérico).
- ✅ **Unificación de respuestas categóricas** (como "yes", "no") a formato binario (1 y 0).
- ✅ **Creación de variable nueva**: `FacturaDiaria`, calculada a partir del valor mensual dividido entre 30.
- ✅ **Renombramiento de columnas clave** para facilitar la comprensión.

---

## 🔹 Análisis Exploratorio de Datos

Se realizaron múltiples análisis para comprender la distribución de evasión:

### 📌 Evasión General
- Aproximadamente **27% de los clientes han cancelado el servicio**.
- Gráfico de barras muestra un desbalance moderado entre clases (`Churn = 0` vs `Churn = 1`).

### 📌 Variables Categóricas
- **Clientes con contrato mes a mes** tienen tasas de cancelación significativamente más altas.
- **Método de pago "Electronic check"** está más asociado a evasión.
- No se observaron diferencias marcadas por **género**.

### 📌 Variables Numéricas
- **Clientes con menos tiempo (`MesesCliente`)** son más propensos a cancelar.
- **FacturaMensual alta** también parece asociarse con mayor probabilidad de cancelación.
- Los clientes que se quedan suelen haber **gastado más en total**.

---

## 🔹 Conclusiones e Insights

- **El tipo de contrato** es el factor más claro asociado a la evasión: los contratos de largo plazo (1 o 2 años) retienen mejor a los clientes.
- **El método de pago y el monto mensual** también son señales de riesgo. Pagos electrónicos y cargos altos están correlacionados con mayor evasión.
- Los **clientes nuevos (poco tiempo)** son el grupo con mayor riesgo de cancelación.

---

## 🔹 Recomendaciones Estratégicas

1. **Incentivar contratos de largo plazo** con descuentos o beneficios exclusivos.
2. **Monitorear clientes con cargos mensuales altos** y ofrecer paquetes ajustados o personalizados.
3. **Ofrecer asistencia proactiva** a nuevos clientes durante sus primeros 3 a 6 meses.
4. **Analizar el método de pago** como un predictor temprano y considerar beneficios por otros métodos más estables.
5. Implementar un **sistema de alertas tempranas** basado en los perfiles de mayor riesgo identificados.
